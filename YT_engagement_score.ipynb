```python
import pandas as pd
import configuration
from googleapiclient.discovery import build
import numpy as np

def engagement_score(video_id):
    video_response = youtube.videos().list(
            id=video_id,
            part='statistics'
        ).execute()
    # Get the statistics for the video.
    statistics = video_response['items'][0]['statistics']
    # Calculate the engagement score.
    views = int(statistics['viewCount'])
    likes = int(statistics.get('likeCount', 0))
    comments = int(statistics.get('commentCount', 0))
    return (views + likes + comments)

config=configuration

API_KEY = config.API_KEY

# Fecthing the INPUT country's list required for slicing the engagement score
country=config.COUNTRY_SLICE
country_list=country.split(',')
print("The engagement score will be calculated for these list of country's =>",country_list)

youtube = build('youtube', 'v3', developerKey=API_KEY)
 
# Read by default 1st sheet and 1st column of an excel file to get the INPUT keywords.
dataframe_keywords = pd.read_excel('/Users/madhusudhan/desktop/youtube/keywords.xlsx',engine='openpyxl')
 
keywords=(dataframe_keywords.iloc[:, 0])

# As per requirement selecting only 100 random keywords.
keywords_random=np.random.choice(keywords, 100, replace=False) 
keyword=""

# Concatinating keywords with "|" for attributing the q for search list API request.

for i in keywords_random:
    keyword=keyword+i+"|"
print("And for the below 100 random list of keyword's\n\n",keyword)

engagement_score_list=[] # List variable to store the engagement scores of all given country .

# Looping over each countr in input list variable country_list.
for country_ in country_list:
    # Search for videos related to the keyword.
    search_response = youtube.search().list(
        q=keyword,
        type='video',
        part='id',
        regionCode=country_,
        maxResults=100
    ).execute()

    # calculate the engagement score based on the number of views, likes, and comments of each video.
    total_engagement = 0
    for search_result in search_response.get('items', []):
        video_id = search_result['id']['videoId']
        engagement_score(video_id)
        total_engagement += engagement_score(video_id)
    engagement_score_list.append(total_engagement)  

# Creating dataframe for storing the engagement score data in a file or console 
ouput_dataframe = pd.DataFrame(list(zip(country_list, engagement_score_list)), columns =['Country', 'Engagement Score']) 

# Printing the output over console in tabular format
print("\n\n",ouput_dataframe.to_markdown(index=False))


```
